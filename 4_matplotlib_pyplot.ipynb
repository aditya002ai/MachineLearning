````python
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
import numpy as np
x = np.arange(0,10)
y = np.arange(11,21)
````

````python
a = np.arange(40,50)
b = np.arange(50,60)
````

````python
plt.scatter(x,y,c='r')
plt.xlabel('X- axis')
plt.ylabel('Y - axis')
plt.title('Graph in 2D')
plt.show()
````

````python
y = x*x
plt.plot(x,y,'r*--',linestyle='dashed', linewidth = 2,markersize = 15)
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
````

````python
## creating sub-plots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'r')
````

````python
x = np.arange(1,11)
y = x*3+5
plt.title('Matplotlib demo')
plt.xlabel('x axis caption')
plt.ylabel('y axis caption')
plt.plot(x,y,'b')
plt.show()
````

````python
np.pi
x=np.arange(0, 4*np.pi,0.1)
y=np.sin(x)
plt.title('sine wave form')
plt.plot(x,y)
plt.show()
````

````python
##Bar plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
````

````python
a = np.array([22,34,56,67,98,90,99,32,45,65,21,12,27,45,25])
plt.hist(a)
plt.title('Histogram')
plt.show()
````

````python
## Box plot using Matlplotlib
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
````

````python
## Pie chart
label = 'python', 'c++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0)
plt.pie(sizes,explode=explode,labels=label,colors=colors,
autopct = '%1.1f%%', shadow = True)
plt.show()
````
